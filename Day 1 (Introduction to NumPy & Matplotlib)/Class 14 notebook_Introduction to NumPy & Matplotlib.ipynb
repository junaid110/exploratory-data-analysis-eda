### Let's  do Comparsion

```python
import time
import numpy as np

# Size of data
N = 10_000_000  

# ----- Python List -----
list_data = list(range(N))

start = time.time()
list_result = [x*2 for x in list_data]   # multiply each element by 2
end = time.time()

print("List time:", end - start, "seconds")

# ----- NumPy Array -----
np_data = np.arange(N)

start = time.time()
np_result = np_data * 2   # vectorized operation
end = time.time()

print("NumPy time:", end - start, "seconds")
```
