## Left Anti Join

![left anti](./img/left_anti_join.png)

```python
left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,2],
              'Units': [40, 25, 30, 35]}

left_table_df = pd.DataFrame(left_table)
left_table_df
```

```python
right_table = {'ID': [3,4],
              'Country': ['Panama', 'Spain']}

right_table_df = pd.DataFrame(right_table)
right_table_df
```

```python
left_anti_join_df = left_table_df.merge(right_table_df,left_on="CountryID",right_on='ID',how='left', indicator=True).drop('ID', axis=1).query('_merge == "left_only"')
```

```python
left_anti_join_df
```

## Right Anti Join

![right anti join](./img/right_anti_join.png)

```python
left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,2],
              'Units': [40, 25, 30, 35]}

left_table_df = pd.DataFrame(left_table)
left_table_df
```

```python
right_table = {'ID': [3,4],
              'Country': ['Panama', 'Spain']}

right_table_df = pd.DataFrame(right_table)
right_table_df
```

```python
right_anti_df = left_table_df.merge(right_table_df,left_on="CountryID", right_on="ID",how="right")
```

```python
right_anti_df = right_anti_df[right_anti_df["CountryID"].isna()].drop(["CountryID", "Date", "Units"], axis=1)   # Method 1
right_anti_df
```

```python
right_anti_join_df = left_table_df.merge(right_table_df,left_on="CountryID",right_on='ID',how='right', indicator=True).drop('ID', axis=1).query('_merge == "right_only"')
# MEthod 2
```

```python
right_anti_join_df
```

```python
right_anti_join_df = left_table_df.merge(right_table_df,
                                   left_on="CountryID",
                                   right_on='ID',
                                   how='right', indicator=True)
```

```python
right_anti_join_df.loc[right_anti_join_df['_merge'] == 'right_only'].drop(["CountryID", "Date", "Units"], axis=1)   # Method 3
```
